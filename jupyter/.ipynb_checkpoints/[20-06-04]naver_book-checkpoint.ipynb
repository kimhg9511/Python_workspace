```python
import requests as req
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
import csv
import json
```

```python
# url - Requests - BeautifulSoup
url = 'https://book.naver.com/search/search.nhn'
params = 'sm=sta_hty.book&sug=&where=nexearch&query=bigdata'
params = list(map(lambda i: i.split('='),params.split('&')))
params = {k: v for k, v in params}
resp = req.get(url, params=params)
soup = bs(resp.text)
```

```python
# find wanted data - title : description
book_all = soup.select_one('#searchBiblioList')
book_all_li = book_all.select('li')
book_all_title = list(map(lambda i : i['alt'], book_all.select('img[onerror]')))
book_all_content = soup.find_all(id=re.compile(r'^searchDescrition'))
remove = { ord('\n'): '', ord('\t'): '', ord('\r'): '', ord('\xa0'): ' : '}
book_all_content = list(map(lambda i: i.text.translate(remove), book_all_content)) 
book_info_all = {k: v for k in book_all_title for v in book_all_content}
# for k, v in book_info_all.items():
#     print(k, v)
```

```python
# author / company / day / price

remove = { ord('\n'): '', ord('\xa0'): '', ord('\t'): '', ord('\r'): ''}
datablocks = soup.select('dd.txt_block')
datablocks = list(map(lambda i: i.text.translate(remove).split('|'), datablocks))
author = []
company = []
date = []
origin_price = []
saled_price = []
ebook_price = []
for datablock in datablocks:
    author.append(' '.join(datablock[0:-2]).strip())
    company.append(datablock[-2].strip())
    date.append(datablock[-1].strip())
datablocks = soup.select('.txt_desc:not(.txt_category)')
datablocks2 = soup.select_one('.txt_desc:not(.txt_category)')
def set_price(elem):
    origin_price = elem.select_one('strike')
    if origin_price != None:
        origin_price = origin_price.text.strip().split('원')[0]
    else:
        origin_price = 0
    saled_price = elem.select_one('em.price')
    if saled_price != None:
        saled_price = saled_price.text.strip().split('원')[0]
    else:
        saled_price = 0
    ebook_price = elem.select_one('img[alt="ebook구매"]')
    if ebook_price != None:
        ebook_price = elem.text.translate(remove).split('|')[-1].split(')')[-1].strip().split('원')[0]
    else:
        ebook_price = 0
#     if elem.select_one('')
    return (origin_price, saled_price, ebook_price)
datablocks = list(map(lambda elem: set_price(elem),datablocks))
for data in datablocks:
    origin_price.append(data[0])
    saled_price.append(data[1])
    ebook_price.append(data[2])
```

```python
# pandas
books = {'author': author,
            'company': company,
            'date': date,
            'origin_price': origin_price,
            'saled_price': saled_price,
            'ebook_price': ebook_price}
books = pd.DataFrame(books)
books
```

```python
# csv
with open('books.csv','w',encoding='utf-8') as books:
    writer = csv.writer(books)
    for i in range(len(author)):
        writer.writerow((author[i], company[i], date[i], origin_price[i], saled_price[i], ebook_price[i]))
with open('books.csv','r',encoding='utf-8') as books:
    reader = csv.reader(books)
    for row in reader:
        print(','.join(row))
    print(reader)
```

```python
# json
json_books = []
with open('books.json','w',encoding='utf-8-sig') as books:
    for i in range(len(author)):
        json_book = {'author': author[i],
                'company': company[i],
                'date': date[i],
                'origin_price': origin_price[i],
                'saled_price': saled_price[i],
                'ebook_price': ebook_price[i]}
        json_books.append(json_book)
    books.write(json.dumps(json_books, ensure_ascii=False, indent=4))
data = []

with open('books.json','r',encoding='utf-8-sig') as books:
    data = json.loads(books.read())
print(data)
```
